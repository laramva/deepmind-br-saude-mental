```python
import pandas as pd

path = "../data/raw/anxiety-disorders-treatment-gap.csv"
df = pd.read_csv(path)


df = df.rename(columns={
    'possibly_adequate_treatment__conditional': 'adequate_treatment',
    'other_treatments__conditional': 'other_treatment',
    'untreated__conditional': 'untreated'
})


for col in ['adequate_treatment', 'other_treatment', 'untreated']:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace("%", "", regex=False)
        .str.replace(",", ".", regex=False)
    )
    df[col] = pd.to_numeric(df[col], errors="coerce")


df.to_csv("../data/processed/anxiety_treatment_gap_processed.csv", index=False)

df.head(), df.columns
```

```python
import pandas as pd


path_raw = "../data/raw/anxiety-disorders-treatment-gap.csv"
df = pd.read_csv(path_raw)

print(df.columns)


df = df.rename(columns={
    "possibly_adequate_treatment__conditional": "adequate_treatment",
    "other_treatments__conditional": "other_treatment",
    "untreated__conditional": "untreated"
})


value_cols = ["adequate_treatment", "other_treatment", "untreated"]

for col in value_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace("%", "", regex=False)   # remove o símbolo de %
        .str.replace(",", ".", regex=False)  # troca vírgula por ponto, se tiver
    )
    df[col] = pd.to_numeric(df[col], errors="coerce")


df_long = df.melt(
    id_vars=["Entity", "Code", "Year"],      # colunas que ficam como "chave"
    value_vars=value_cols,                  # colunas que viram linhas
    var_name="treatment_type",              # nome da coluna que indica o tipo
    value_name="value"                      # nome da coluna numérica
)


output_path = "../data/processed/anxiety_treatment_gap_long.csv"
df_long.to_csv(output_path, index=False)

output_path, df_long.head()
```

```python
import pandas as pd


path = "../data/raw/anxiety-disorders-treatment-gap.csv"
df = pd.read_csv(path)

print("COLUNAS ORIGINAIS:")
print(df.columns.tolist())


adequate_col   = [c for c in df.columns if "adequate" in c.lower()][0]
other_col      = [c for c in df.columns if "other_treat" in c.lower()][0]
untreated_col  = [c for c in df.columns if "untreated" in c.lower()][0]

print("\nColunas detectadas:")
print("Adequate  →", adequate_col)
print("Other     →", other_col)
print("Untreated →", untreated_col)


df = df.rename(columns={
    adequate_col: "adequate",
    other_col: "other",
    untreated_col: "untreated"
})


df = df[["Entity", "Code", "Year", "adequate", "other", "untreated"]]


for col in ["adequate", "other", "untreated"]:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.replace("%", "", regex=False)
    df[col] = df[col].str.replace(",", ".", regex=False)
    df[col] = pd.to_numeric(df[col], errors="coerce")

print("\nPRÉVIA DO DF LIMPO:")
print(df.head())
print(df.dtypes)


df_long = df.melt(
    id_vars=["Entity", "Code", "Year"],
    value_vars=["adequate", "other", "untreated"],
    var_name="Treatment_Type",
    value_name="Value"
)

print("\nVALORES NULOS EM VALUE:", df_long["Value"].isna().sum())
print(df_long.head())


output_path = "../data/processed/anxiety_treatment_gap_long.csv"
df_long.to_csv(output_path, index=False)
print("\nArquivo salvo em:", output_path)
```
