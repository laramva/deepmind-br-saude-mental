```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv(
    "../data/raw/internacoes_mentais.csv",
    sep=';',
    encoding='latin-1',
    skiprows=4
)

df.head()
```

```python
col_anos = [c for c in df.columns if c.isdigit()]
col_anos
```

```python
for col in col_anos:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
internacoes_ano = df[col_anos].sum()
internacoes_ano
```

```python
plt.figure(figsize=(12,4))
internacoes_ano.plot(kind='bar')
plt.title("Internações por transtornos mentais (2008–2025)")
plt.xlabel("Ano")
plt.ylabel("Total")
plt.show()
```
